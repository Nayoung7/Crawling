### 네이버 영화 랭킹(현재상영영화)

```python
# title 변수에 영화제목
# score 변수에 평점

url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date=20220106'
res = req.get(url)
res
```

```python
soup = bs(res.text, 'lxml')
title = soup.select('.tit5 > a')
len(title)
```

```python
score = soup.select('.point')
len(score)
```

```python
title_list=[]
score_list=[]
rank_list=[]

for i in range(len(title)):
    title_list.append(title[i].text)
    score_list.append(score[i].text)
    rank_list.append(i + 1)
```

```python
dic={'영화제목':title_list, '평점':score_list, '순위':rank_list}
```

```python
import pandas as pd
df = pd.DataFrame(dic)
df.set_index('순위',inplace=True)
df
```

### 한달동안의 영화데이터 수집

```python
'안'+'녕하세요'
```

```python
1월 1일
20220101~20220106
```

```python
title_list=[]
score_list=[]
rank_list=[]

for i in range(20220101,20220107):
    url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date='+ str(i)

    res = req.get(url)
    res

    soup = bs(res.text,'lxml')
    title = soup.select('div.tit5 > a')
    score = soup.select('td.point')

    for i in range(len(title)):
        title_list.append(title[i].text)
        score_list.append(score[i].text)
        rank_list.append(i + 1)
```

```python
len(title_list)
```

```python
dic = { '영화제목' : title_list, '평점' : score_list, '순위' : rank_list}

df = pd.DataFrame(dic)
df.set_index('순위', inplace = True)
```

### Pandas 활용하여 날짜 생성

```python
data = pd.date_range(start='2021-01-01', end='2021-03-31')
data
```

```python
# 문자형 포맷팅

days = data.strftime('%Y%m%d')
days
```

```python
from tqdm import tqdm_notebook
```

```python
title_list=[]
score_list=[]
rank_list=[]
day_list=[]

for day in tqdm_notebook(days):
    url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date='+ day

    res = req.get(url)
    res

    soup = bs(res.text,'lxml')
    title = soup.select('div.tit5 > a')
    score = soup.select('td.point')

    for i in range(len(title)):
        title_list.append(title[i].text)
        score_list.append(score[i].text)
        rank_list.append(i+1)
        day_list.append(day)
```

```python
dic = { '영화제목' : title_list, '평점' : score_list, '순위' : rank_list, '날짜':day_list}

df = pd.DataFrame(dic)
df.set_index('순위', inplace = True)
```
